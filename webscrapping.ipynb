```python
import requests
from bs4 import BeautifulSoup

url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

response = requests.get(url)

print(response.status_code)


soup=BeautifulSoup(response.text,"lxml")

prices=soup.find_all("h4",class_="price float-end card-title pull-right")

print(f"""Total items {len(prices)}""")

print(f"""Price at index {prices[3].text}""")

for i in prices:
    print(i.text)

```

# Pandas in scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

response = requests.get(url)

print(response.status_code)

soup=BeautifulSoup(response.text,"lxml")

names=soup.find_all("a",class_="title")
product_name_list=[]

for i in names:
    name=i.text
    product_name_list.append(name)
    
print(product_name_list)


prices=soup.find_all("h4",class_="price float-end card-title pull-right")
product_price_list=[]

for i in prices:
    price=i.text
    product_price_list.append(price)
    
print(product_price_list)

descriptions=soup.find_all("p",class_="description card-text")
product_description_list=[]

for i in descriptions:
    description=i.text
    product_description_list.append(description)

print(product_description_list)

reviews=soup.find_all("p",class_="review-count float-end")
product_review_list=[]

for i in reviews:
    review=i.text
    product_review_list.append(review)

print(product_review_list)

df=pd.DataFrame({"Product Name":product_name_list,"Product Decriiption":product_description_list,"Product Price":product_price_list,"Product Reviews":product_review_list})
df.to_csv("Product_detail.csv")




```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'

# Send a GET request to fetch the page content
response = requests.get(url)

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find the first table containing the GDP data
table = soup.find('table', {'class': 'wikitable'})

if table:
    # List to store the extracted data
    all_data = []

    # Manually set the column headers since the table has more than 4 columns
    columns = [
        'Country/Territory', 
        'IMF Forecast (2024)', 'IMF Year', 
        'World Bank Estimate (2023)', 'World Bank Year', 
        'United Nations Estimate (2022)', 'UN Year'
    ]

    # print(f"Columns: {columns}")

    # Extract the table rows (skip the header)
    rows = table.find_all('tr')[1:]

    # Loop through the rows to extract the country GDP data
    for index, row in enumerate(rows):
        cells = row.find_all(['td', 'th'])
        data_row = [cell.get_text(strip=True).replace('\n', ' ') for cell in cells]
        
        # Debugging - Print each row's content
        print(f"Row {index}: {data_row}")
        
        # Ensure only rows with the expected number of columns (7) are processed
        if len(data_row) == 7:
            all_data.append(data_row)
        else:
            print(f"Skipping row {index} due to column mismatch (Expected 7 columns, got {len(data_row)})")
    
    # Debugging - Print all the collected data
    print(f"All data collected: {all_data}")

    # Check if any data was collected
    if all_data:
        # Create a DataFrame using the collected data
        df = pd.DataFrame(all_data, columns=columns)

        # Display the DataFrame
        print(df)

        # Save the data to a CSV file
        df.to_csv('countries_by_gdp.csv', index=True)
        print("Data has been successfully saved to 'countries_by_gdp.csv'")
    else:
        print("No valid data rows were found.")
else:
    print("Table not found on the Wikipedia page.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the CoinMarketCap page
base_url = 'https://coinmarketcap.com/'

# List to store all data
all_data = []
columns = []

# Loop through pagination (1 to 5 pages as an example, adjust range as needed)
for page in range(1, 6):  # Adjust the range to scrape more pages
    # Construct the URL with the current page number
    url = f'{base_url}?page={page}'
    
    # Send a GET request to fetch the page content
    response = requests.get(url)
    
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table that contains the cryptocurrency data
    table = soup.find('table')
    
    # Find the table headers if it's the first page (to dynamically get the column names)
    if page == 1:
        header_row = table.find('thead').find_all('th')
        columns = [th.text.strip() for th in header_row]
    
    # Get all the rows from the table
    rows = table.find('tbody').find_all('tr')
    
    # Loop through rows to extract currency data
    for row in rows:
        data_row = [td.text.strip() for td in row.find_all('td')]
        if data_row:
            all_data.append(data_row)

# Create a DataFrame using the collected data and dynamic columns
df = pd.DataFrame(all_data, columns=columns)

# Display the DataFrame
print(df)

# Optionally, save the data to a CSV file
df.to_csv('coinmarketcap_all_data.csv')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Etherscan Tokens page
url = "https://etherscan.io/tokens"

# Send a GET request to fetch the page content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the table containing token data
    table = soup.find('table', {'class': 'table table-hover'})

    # Check if the table exists
    if table:
        # Extract table headers
        headers = [th.text.strip() for th in table.find_all('th')]

        # Extract table rows
        rows = []
        for tr in table.find_all('tr')[1:]:  # Skipping header row
            cells = [td.text.strip() for td in tr.find_all('td')]
            if cells:
                rows.append(cells)

        # Create a DataFrame using pandas
        df = pd.DataFrame(rows, columns=headers)

        # Display the DataFrame
        print(df)

        # Optionally save to CSV
        df.to_csv('etherscan_tokens.csv', index=False)
        print("Data saved to 'etherscan_tokens.csv'")
    else:
        print("Table not found on the page.")
else:
    print(f"Failed to retrieve page. Status code: {response.status_code}")
```
